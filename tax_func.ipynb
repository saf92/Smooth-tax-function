```python
x = np.linspace(10**2,2*10**5,10**4)
x_t = income_after_tax1(x)

# Given constant
p = 0.53

# Define the model function
def model1(x, gamma):
    return income_after_tax_func1(x,gamma,p)

# Fit the model to the data
initial_guess1 = [10**-5]  # starting guess for gamma
params1, covariance1 = curve_fit(model1, x, x_t, p0=initial_guess1)

# Extract fitted gamma
gamma_fit1 = params1[0]
gamma_err1 = np.sqrt(np.diag(covariance1))[0]

print(gamma_fit1,gamma_err1)

x_t_pred1 = model1(x,gamma_fit1)
```

```python
def model2(x, a, gamma):
    # Essentially same as model 1!
    return p*x+((1-p)*a**gamma)/(1-gamma)*((x+a)**(1-gamma)-a**(1-gamma))

# Fit the model to the data
initial_guess2 = [0.1,0.1]  # starting guess 
params2, covariance2 = curve_fit(model2, x, x_t, p0=initial_guess2)

# Extract fitted params
params_fit2 = params2[0]
params_err2 = np.sqrt(np.diag(covariance2))[0]

print(params_fit2,params_err2)

x_t_pred2 = model2(x,*params2)
```

```python
def sum_squares(x,y):
    n = len(x)
    return 1/n*np.sum((x-y)**2)

print(sum_squares(x_t,x_t_pred1),sum_squares(x_t,x_t_pred2))
```
